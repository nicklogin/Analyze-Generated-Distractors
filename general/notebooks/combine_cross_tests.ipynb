```python
import pandas as pd
```

```python
cross_test_ege = pd.read_excel("../data_for_comparison/cross_tests/CrossTestEge.xlsx")
cross_test_ege["Датасет"] = "ЕГЭ"
cross_test_ru_race_tf = pd.read_csv("../data_for_comparison/cross_tests/cross_test_ru_race_tf.csv", index_col="Unnamed: 0")
```

```python
cross_test_ege
```

```python
cross_test_ru_race_tf = cross_test_ru_race_tf.rename(
    columns={
        "vso_intersec_ind": "FullVPMatch",
        "vs_intersec_ind": "SubjVPMatch",
        "noun_intersec_ind": "NounMatch",
        "propn_intersec_ind": "PropnMatch"
    }
).drop("vs_passivized_intersec_ind", axis="columns")
cross_test_ru_race_tf["Датасет"] = "Ru-RACE-TF"
cross_test_ru_race_tf["FullVPMatch"] *= 100
cross_test_ru_race_tf["SubjVPMatch"] *= 100
cross_test_ru_race_tf["NounMatch"] *= 100
cross_test_ru_race_tf["PropnMatch"] *= 100
cross_test_ru_race_tf
```

```python
df_compare = pd.concat([cross_test_ege, cross_test_ru_race_tf])
df_compare = df_compare.set_index(["Датасет", "Источник"])
df_compare
```

```python
df_compare.to_excel("../data_for_comparison/cross_tests/cross_tests.xlsx", float_format="%.2f")
```
