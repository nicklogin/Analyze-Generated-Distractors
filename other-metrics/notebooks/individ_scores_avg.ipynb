```python
import pandas as pd
```

```python
def read_excel_multisheet(filename: str) -> pd.DataFrame:
    workbook = pd.ExcelFile(filename)
    sheets = workbook.sheet_names
    df = pd.concat([pd.read_excel(workbook, sheet_name=s, index_col="Unnamed: 0").sort_index()
                    .assign(source=s) for s in sheets])
    return df
```

```python
individ_scores = read_excel_multisheet("../data_for_comparison/avg/individ_metrics.xlsx")
```

```python
table1 = individ_scores.drop(
    ["qid", "options", "ground_truth", "common_ngrams"], axis="columns"
).groupby("source").mean()
```
